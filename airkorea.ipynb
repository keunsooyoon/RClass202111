```r
#  전국의 대기오염 측정 자료 

files <- c('ds.2015.csv','ds.2016.csv','ds.2017.csv','ds.2018.csv','ds.2019.csv')
```

```r
ds <- NULL

for (f in files) {
    tmp <- read.csv(f, header=T)
    ds <- rbind(ds, tmp)
    print(f)
}
```

```r
ds$locname[ds$loc==111123] <- '서울'
ds$locname[ds$loc==336111] <- '목포'
ds$locname[ds$loc==632132] <- '강릉'
```

```r
head(ds,1)
```

```r
#연도 월 시간 대로 그룹핑을하여 분석하기 쉽게 파생변수 만들기

mdate <- as.character(ds$mdate)
```

```r
ds$year <- as.numeric(substr(mdate,1,4))  # 연도 추출
ds$month <- as.numeric(substr(mdate,5,6))  # 연도 추출
ds$hour <- as.numeric(substr(mdate,9,10))  # 연도 추출
```

```r
head(ds,1)
```

```r
# 지역별 PM10 농도 분석

boxplot(PM10~locname, data=ds, main='미세먼지 농도 분포')
```

```r
boxplot(PM10~locname, data=ds, 
            main='미세먼지 농도 분포',
            ylim=c(1,100))
# 세 지역 모두 정상 범위 보다 미세 먼지 농도가 극단적으로 높아지는 시점들이
# 존재한다. 특히 서울 과 강릉 지역이 더 심하다. 
# 세지역 모두 정상 범위는 30-50사이로 비슷하다. 
```

```r
# 연도별 PM10 농도 추이

tmp.year <- aggregate(ds[,7], 
                      by=list(year=ds$year, loc=ds$locname), FUN='mean')
tmp.year$loc = as.factor(tmp.year$loc)
head(tmp.year)
```

```r
library(ggplot2)
ggplot(tmp.year, aes(x=year, y=x, colour=loc, group=loc)) +
        geom_line() +
        geom_point(size=6, shape=19, alpha=0.5) +
        ggtitle('연도별 PM10 농도 변화') +
        ylab('농도')
```

```r
head(ds,1)
```

```r
tmp.month <- aggregate(ds[,7], 
                       by=list(month=ds$month, loc=ds$locname),FUN='mean')
tmp.month
```

```r
tmp.month$loc = as.factor(tmp.month$loc)
```

```r
ggplot(tmp.month, aes(x=month, y=x, colour=loc, group=loc))+
    geom_line() +
    geom_point(size=3, shape=19, alpha=0.5) +
    ggtitle('월별 PM10 농도 변화') +
    ylab('농도')

# 월별 미세먼지 농도 변화가 겨울에서 봄으로 가면서 높아지다가 장마철인
# 여름에는 미세먼지 농도가 낮아지는 것을 볼 수 있다. 
# 서울 지역은 겨울과 봄에는 미세먼지 농도가 가장 높은데
# 오히려 여름에는 미세먼지 농도가 가장 낮아지는 것을 볼수 있다. 
```

```r
# 시간대별 PM10 농도 추이
```

```r
tmp.hour <- aggregate(ds[,7], 
                      by=list(hour=ds$hour, loc=ds$locname),FUN='mean')
```

```r
tmp.hour$loc = as.factor(tmp.hour$loc)
head(tmp.hour)
```

```r
ggplot(tmp.hour, aes(x=hour, y=x, colour=loc, group=loc)) +
    geom_line()+
    geom_point(size=3, shape=19, alpha=0.5) +
    ggtitle('시간별 PM10 농도 변화') +
    ylab('농도')
```

```r
# 시간대별 미세먼지 농도 추이는 아침 시간대는 낮아 졌다가 서서히 
# 올라가서 져녁시간대에 가장 높아지는 것을 볼 수 있다. 
# 강릉의 경우 시간대별 미세먼지 농도 변화의 폭이 매우 크다. 
```
