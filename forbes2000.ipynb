```r
install.packages("HSAUR")
library(HSAUR)

data("Forbes2000")
ds <- Forbes2000


head(ds)
tail(ds)
dim(ds)
str(ds)
summary(ds)

# rank - 기업 순위, name - 기업 이름, country - 소속 국가
# category - 업종 구분, sales - 매출액, profits - 순익
# assets - 자산 marketvalue - 기업 가치


#결측치 확인
ds[!complete.cases(ds),]


# 국가별 기업 통계
table(ds$country)

# 내림차순으로 정렬
tmp <- sort(table(ds$country), decreasing = T)
tmp

# 상위 10개국 
top.10.country <- tmp[1:10]
top.10.country

# 그래프 출력
barplot(top.10.country,
        main='기업수 상위 10개국')


barplot(top.10.country,
        main='기업수 상위 10개국',
        col=rainbow(10),
        las=2)

# 업종별 기업 통계

table(ds$category)
tmp <- sort(table(ds$category), decreasing=T)
tmp

top.10.category <- tmp[1:10]
top.10.category

barplot(top.10.category,
        main="top 10 category")

barplot(top.10.category,
        main="top 10 category",
        col=rainbow(10),
        las=2)
```

```r
# 업종별 기업자산 분포

tmp <- ds[ds$category %in% names(top.10.category),]
tmp

# A %in% B -> A에 있는 값들 중 B에 속하는 값들을 찾는 역할

# category  열의 값이 팩터 타입 tmp$category <- factor(tmp$category)로 
# 설정이 필요
levels(tmp$category)
```

```r
tmp$category <-factor(tmp$category)
```

```r
levels(tmp$category)
```

```r
boxplot(assets~category, data=tmp, las=2, xlab='', ylim=c(0,100))
#업종별 기업 자산 부포 역시 은행과 보험 업종 기업 자산이 
#많은 것을 볼 수 있다. 
```

```r
# 기업 가치 상위 10대 기업

tmp <- ds[order(ds$marketvalue, decreasing=T),]
tmp[1:10,c('name','country','category','marketvalue')]
```

```r
# 한국 기업 정보
korea <- subset(ds, country=='South Korea')
korea[,c('rank','name','category','marketvalue')]
```

```r
# 기업 가치 와 타 변수와의 상관관계 
tmp <- ds[,5:8]
tmp <- tmp[complete.cases(tmp),]
tmp
```
